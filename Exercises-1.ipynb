```python
import pandas as pd
```

```python
cast = pd.read_csv('E:/1 Top Mentor Data Analyst Classes/Data sets/cast.csv')
cast.head()
```

### How many movies are listed in the titles dataframe?

```python
total_movies = cast.shape[0]
print(f"Total number of titles: {total_movies}")
```

### In what years has a movie titled "Batman" been released?

### How many roles were there in the movie "Inception"?

```python
inception_roles = cast[cast['title'] == 'Inception']
role_count = inception_roles.shape[0]
print(f"Number of roles in 'Inception': {role_count}")
```

### How many roles in the movie "Inception" are NOT ranked by an "n" value?

```python
unranked_roles = inception_roles['n'].isna().sum()
print(f"Number of unranked roles in 'Inception': {unranked_roles}")
```

### But how many roles in the movie "Inception" did receive an "n" value?

```python
ranked_roles = inception_roles['n'].notna().sum()
print(f"Number of ranked roles in 'Inception': {ranked_roles}")
```

### Display the cast of "North by Northwest" in their correct "n"-value order, ignoring roles that did not earn a numeric "n" value.

```python
nbnw_cast = cast[(cast['title'] == 'North by Northwest') & (cast['n'].notna())]
# Sort by 'n' value
nbnw_cast_sorted = nbnw_cast.sort_values('n')
# Display name, character, and n
print(nbnw_cast_sorted[['name', 'character', 'n']].to_string(index=False))
```

### Display the entire cast, in "n"-order, of the 1972 film "Sleuth".

```python
sleuth_cast = cast[(cast['title'] == 'Sleuth') & (cast['year'] == 1972) & (cast['n'].notna())]
# Sort by 'n' value
sleuth_cast_sorted = sleuth_cast.sort_values('n')
# Display name, character, and n
print(sleuth_cast_sorted[['name', 'character', 'n']].to_string(index=False))
```

### Now display the entire cast, in "n"-order, of the 2007 version of "Sleuth".

### List the supporting roles (having n=2) played by Cary Grant in the 1940s, in order by year.

```python
grant_supporting_40s = cast[
    (cast['name'] == 'Cary Grant') &
    (cast['n'] == 2) &
    (cast['year'] >= 1940) &
    (cast['year'] <= 1949)
]

# Sort by year
grant_supporting_40s_sorted = grant_supporting_40s.sort_values('year')

# Display title, year, and character
print(grant_supporting_40s_sorted[['title', 'year', 'character']].to_string(index=False))
```

### List the leading roles that Cary Grant played in the 1940s in order by year.

```python
grant_leading_40s = cast[
    (cast['name'] == 'Cary Grant') &
    (cast['n'] == 1) &
    (cast['year'] >= 1940) &
    (cast['year'] <= 1949)
]

# Sort by year
grant_leading_40s_sorted = grant_leading_40s.sort_values('year')

# Display title, year, and character
print(grant_leading_40s_sorted[['title', 'year', 'character']].to_string(index=False))
```

### How many roles were available for actors in the 1950s?

### How many leading roles (n=1) were available from the beginning of film history through 1980?

```python
leading_roles = cast[
    (cast['n'] == 1) &
    (cast['year'] <= 1980)
]

# Count total number of leading roles
total_leading_roles = leading_roles.shape[0]

print(f"Total number of leading roles through 1980: {total_leading_roles}")
```

### How many non-leading roles were available through from the beginning of film history through 1980?

```python
non_leading_roles = cast[
    (cast['n'] != 1) &
    (cast['year'] <= 1980)
]

# Count total number of non-leading roles
total_non_leading_roles = non_leading_roles.shape[0]

print(f"Total number of non-leading roles through 1980: {total_non_leading_roles}")
```


### How many roles through 1980 were minor enough that they did not warrant a numeric "n" rank?
