### Step 9. Which was the most-ordered item?

```python
top_item = chipo.groupby('item_name')['quantity'].sum().idxmax()
print(top_item)
```

### Step 10. For the most-ordered item, how many items were ordered?

```python
item_totals = chipo.groupby('item_name')['quantity'].sum()
top_item_name = item_totals.idxmax()
top_item_quantity = item_totals.max()
print(f"{top_item_name}: {top_item_quantity} items ordered")
```

### Step 11. What was the most ordered item in the choice_description column?

```python
choice_totals = chipo.groupby('choice_description')['quantity'].sum()
# Find the most ordered choice_description
top_choice = choice_totals.idxmax()
top_quantity = choice_totals.max()
print(f"{top_choice}: {top_quantity} items ordered")
```

### Step 12. How many items were orderd in total?

```python
total_items = chipo['quantity'].sum()
print(f"Total items ordered: {total_items}")
```

### Step 13. Turn the item price into a float

#### Step 13.a. Check the item price type

```python
print(chipo['item_price'].dtype)
```

#### Step 13.b. Create a lambda function and change the type of item price

```python
chipo['item_price'] = chipo['item_price'].apply(
    lambda x: float(x.replace('$', '')) if isinstance(x, str) else x
)
print(chipo['item_price'].head())
```

#### Step 13.c. Check the item price type

```python
print(chipo['item_price'].dtype)
```

### Step 14. How much was the revenue for the period in the dataset?

```python
chipo['revenue'] = chipo['quantity'] * chipo['item_price']
total_revenue = chipo['revenue'].sum()
print(f"Total revenue: ${total_revenue:.2f}")
```

### Step 15. How many orders were made in the period?

```python
total_orders = chipo['order_id'].nunique()
print(f"Total number of orders: {total_orders}")
```

### Step 16. What is the average revenue amount per order?

```python
# Solution 1
average_revenue = total_revenue / total_orders

print(f"Average revenue per order: ${average_revenue:.2f}")
```

### Step 17. How many different items are sold?

```python
unique_items = chipo['item_name'].nunique()

print(f"Number of different items sold: {unique_items}")
```
